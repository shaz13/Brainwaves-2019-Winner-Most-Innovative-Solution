```python
import csv
import glob
import requests 
import pandas as pd
import numpy as np
import xml.etree.ElementTree as ET 
from bs4 import BeautifulSoup
from tqdm import tqdm


train_paths =  []
for filename in glob.iglob('../data/train_data/*xml', recursive=True):
    train_paths.append(filename)

test_paths =  []
for filename in glob.iglob('../data/private_test/*xml', recursive=True):
    test_paths.append(filename)
```

```python
train_filenames = [x.split("/")[-1].split('.')[0] for x in train_paths]
test_filenames = [x.split("/")[-1].split('.')[0] for x in test_paths]
```

```python
train_labels = pd.read_csv("../data/adhoc/train_labels.csv")
```

```python
test_baseline = pd.DataFrame(columns=train_labels.columns)
```

```python
test_baseline['File_Name'] = test_filenames
```

```python
for col in train_labels.columns[2:]:
    test_baseline[col] = train_labels[col].value_counts().index[0]
```

```python
test_baseline.to_csv("./baseline.csv", index=None)
```
